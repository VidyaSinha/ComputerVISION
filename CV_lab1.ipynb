```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img= cv2.imread("/content/images.jpg") #gives rgb
imgG= cv2.imread("/content/images.jpg",0) #gives grayscale

cv2_imshow(img)
```

```python
cv2_imshow(imgG)
```

```python
# 2 types of resize:
# 1.resize within the pixel ratio
# 2. Resize in percent ratio

img=cv2.resize(imgG,(400,150))
imgG=cv2.resize(imgG,(0,0),fx=0.3,fy=0.2)
cv2_imshow(imgG)
cv2_imshow(img)
```

```python
#rotation
imgR= cv2.rotate(imgG,cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(imgR)
```

```python
img= cv2.line(img,(0,0),(img.shape[1],img.shape[0]),(255,0,0),6)
cv2_imshow(img)
```

```python
img = cv2.line(img, (img.shape[1] - 1, 0), (0, img.shape[0] - 1), (255, 0, 0), 6)
cv2_imshow(img)
```

```python
img=cv2.rectangle(img,(80,40),(40,60),(0,0,255),-1)
cv2_imshow(img)
```

```python
#write the text
font_stye= cv2.FONT_HERSHEY_SIMPLEX
img=cv2.putText(img,'Hello ICTians !!',(10,img.shape[0]-10),font_stye,1,(0,0,0),2)
cv2_imshow(img)
```
