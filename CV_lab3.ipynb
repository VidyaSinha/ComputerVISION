```python
#masking of image

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('/content/sunflower.jpg')
cv2_imshow(img)
img.shape
```

```python
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)
```

```python
#mask of green color
mask=cv2.inRange(hsv,(36,0,0),(70,255,255))
bitand1= cv2.bitwise_and(img,img,mask=mask)
cv2_imshow(bitand1)
```

```python
# mask of yellow colour
mask1=cv2.inRange(hsv,(15,0,0),(36,255,255))
bitand1= cv2.bitwise_and(img,img,mask=mask1)
cv2_imshow(bitand1)
```

```python
bitor=cv2.bitwise_or(mask,mask1)
final=cv2.bitwise_and(img,img,mask=bitor)

```

```python
cv2_imshow(img)
print()
cv2_imshow(final)
```
