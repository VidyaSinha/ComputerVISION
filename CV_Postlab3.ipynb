```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('/content/shoes.jpg')
img=cv2.resize(img,(300,400))
cv2_imshow(img)
```

```python
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)
```

```python
#mask of green color
mask=cv2.inRange(hsv,(36,0,0),(70,255,255))
bitand1= cv2.bitwise_and(img,img,mask=mask)
cv2_imshow(bitand1)
```

```python
# mask of red colour
mask1=cv2.inRange(hsv,(170,50,50),(180,255,255))
bitand1= cv2.bitwise_and(img,img,mask=mask1)
cv2_imshow(bitand1)
```

```python
bitor=cv2.bitwise_or(mask,mask1)
final=cv2.bitwise_and(img,img,mask=bitor)
```

```python
# Convert to HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# Mask for green color range (background example)
mask_green = cv2.inRange(hsv, (36, 0, 0), (70, 255, 255))

# Mask for red color range (example for combining foreground ranges)
mask_red1 = cv2.inRange(hsv, (0, 50, 50), (10, 255, 255))
mask_red2 = cv2.inRange(hsv, (170, 50, 50), (180, 255, 255))
mask_red = cv2.bitwise_or(mask_red1, mask_red2)

# Combine masks for all foreground colors
foreground_mask = cv2.bitwise_or(mask_green, mask_red)

# Invert the mask to isolate the background
background_mask = cv2.bitwise_not(foreground_mask)

# Extract the background
background_only = cv2.bitwise_and(img, img, mask=background_mask)

# Show results
cv2_imshow(background_only)
```
